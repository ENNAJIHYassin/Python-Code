## Brown Corpus

```python
# Import the brown corpus
from nltk.corpus import brown

# Build the conditional Frequency Distribution
cfd = nltk.ConditionalFreqDist((genre, words)
                              for genre in brown.categories()
                              for words in brown.words(categories = genre))

# Enonciate our modals
modals = ['can', 'could', 'may', 'might', 'must', 'will']

# Draw the table of conditional frequency by using the modals
cfd.tabulate(conditions = brown.categories(), samples = modals)

#########################################################################################
### The Brown Corpus is a convenient resource for studying systematic differences between
### genres, a kind of linguistic inquiry known as stylistics
#########################################################################################
### Word counts might distinguish genres !!!!!
#########################################################################################
```

## Inaugural Address Corpus

```python
# Import the Inaugural Corpus
from nltk.corpus import inaugural

# Build the conditional Frequency Distribution
cfd = nltk.ConditionalFreqDist( (target, fileid[:4])
                              for fileid in inaugural.fileids()
                              for w in inaugural.words(fileid)
                              for target in ["america", "citizen"]
                              if w.lower().startswith(target))

# Plot the result
cfd.plot() #Keep in mind that counts are not normalized for documents lenght
```

# 2.2 Conditional Frequency Distributions

## Counting Words by Genre

```python
# Construct pairs of words (genre, word) to feed the conditional frequency function 
from nltk.corpus import brown
genre_word = [(genre, word)
             for genre in ["news", "romance"]
             for word in brown.words(categories=genre)]
len(genre_word)
```

```python
# Pairs at the beginning of the list genre_word will be of the form ('news', word)
genre_word[:4]
```

```python
# Whereas those at the end will be of the form ('romance', word)
genre_word[-4:]
```

```python
# Create the conditional frequency distribution
cfd = nltk.ConditionalFreqDist(genre_word)
type(cfd)
```

```python
# Access the conditions
cfd.conditions()
```

```python
# Access the freqdist of a specefic condition
cfd['news']
```

```python
# Access the words within a condition
list(cfd['romance'])
```

```python
# Access the frequency of a specefic word within a condition
cfd["romance"]["could"]
```

## Plotting and Tabulating Distributions

```python
# Import corpus and define languages
from nltk.corpus import udhr
languages = ['Chickasaw', 'English', 'German_Deutsch','Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']

# Create conditional distribution
cfd = nltk.ConditionalFreqDist((lang, len(word))
                               for lang in languages
                               for word in udhr.words(lang + '-Latin1'))

# We interpret the last cell on the top row to mean that 1,638 words of the English text have nine or fewer letters.
cfd.tabulate(conditions=['English', 'German_Deutsch'], 
             samples=range(10),
             cumulative=True)
```

## Generating Random Text with Bigrams

```python
# Random text generation function
def generate_text(cfdist, word, num=15):
    for _ in range(num):
        print(word, end=" ")
        word = cfdist[word].max()
        
text = nltk.corpus.genesis.words("english-kjv.txt")
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)

generate_text(cfd, "living")
```

```python
# Function to compute what fraction of words in a text are not in the stopwords list
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content)/len(text)

content_fraction(nltk.corpus.reuters.words())
```

```python
# Solve the word puzzle "Target"
puzzle_letters = nltk.FreqDist('CAEOTRMSH'.lower()) 
obligatory = 't'
puzzle_letters
```

```python
eng_words = nltk.corpus.words.words()
solution = [w for w in eng_words if len(w) >= 4
           and obligatory in w
           and nltk.FreqDist(w) <= puzzle_letters]
solution
```

```python
entries = nltk.corpus.cmudict.entries()
len(entries)
```

```python
# Look for special words' phonetics
for word, pron in entries:
    if len(pron) == 3:
        ph1, ph2, ph3 = pron
        if ph1 =='P' and ph3 =='T':
            print(word, ph2, end=" ")
```

```python
# Find words that rhym with "nicks"
syllable = ['N', 'IH0', 'K', 'S']
[w for w, pron in entries if pron[-4:] == syllable]
```

```python
# Look for words that ending with 'M' sound but end with 'n' letter
[w for w,pron in entries if pron[-1] == 'M' and w[-1]=='n']
```

```python
# Looking for letter combination in words starting with 'N' sound but don't begin with the letter 'n'
set(w[:2] for w,pron in entries if pron[0]=='N' and w[0]!='n')
```

```python
def stress(pron):
    return [char for phone in pron 
                 for char in phone if char.isdigit()]
```

```python
[w for w, pron in entries if stress(pron) == ['0', '1', '0', '2', '0']]
```

```python
[w for w, pron in entries if stress(pron) == ['0', '2', '0', '1', '0']]
```

```python
p3 = [(pron[0]+'-'+pron[2], word)
    for (word, pron) in entries
    if pron[0] == 'P' and len(pron) == 3]

cfd = nltk.ConditionalFreqDist(p3)
#cfd.conditions()
cfd['P-D']
for template in cfd.conditions():
    if len(cfd[template]) > 10:
        words = list(cfd[template].keys())
        wordlist = ' '.join(words)
        print(template, wordlist[:70] + "...")
        
```
