## Counting Vocabulary

```python
# Alphabetical order of unique "types" in a text
sorted(set(text3))  # Capitalized words appear before lowercase words in sorted lists
```

```python
#######################################
### THREE MAINS ATTRIBUTES OF A TEXT :
  #  Tokens
  #  Types
  #  Lexical Diversity
#######################################

# Number of tokens in a text
def tokens(text):
    return len(text)

# Number of unique types in text (type = word + punctuation and special characters)
def types(text):
    return len(set(text))

# Number of times each unique type is used in a text on average
def lexical_diversity(text):
    return tokens(text) / types(text)

################################################
### TWO MAINS ATTRIBUTES OF A WORD IN A TEXT :
  #  Count
  #  Percentage
###############################################
    
# Number of times a word is used in a text
def count(word, text):
    return text.count(word)

# Percentage of the text is taken up by a specific word
def percentage(word, text):
    return 100 * count(word, text) / tokens(text)
```

## Counting Other Things

```python
# The distribution of word lengths in a text
fdist = FreqDist([len(w) for w in text1])
list(fdist.keys()) # the lengh of every single word in text1 is in this list
```

```python
# Frequency of the different lengths of words
sorted(list(fdist.items()), key = lambda wlf : wlf[1], reverse = True)
```

```python
# Get the word lengh that appears the most
fdist.max()
```

```python
# Get the frequency of that word
fdist[fdist.max()]
```

```python
# Get the percentage of that word
fdist.freq(fdist.max()) # words of length 3 account for roughly 50,000 (or 20%) of the words making up the book
                        # further analysis of word length might help us understand differences between authors, genres, or languages
```
